### Product Name vs Order 

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Sample Dataset.xlsx'


orders_df = pd.read_excel(file_path, sheet_name='Sales Orders Sheet')
products_df = pd.read_excel(file_path, sheet_name='Products Sheet')


merged_df = pd.merge(orders_df, products_df, on='Product Code', how='left')

product_order_summary = merged_df.groupby('Product Name')['Order Quantity'].sum().reset_index()

product_order_summary = product_order_summary.sort_values(by='Order Quantity', ascending=False)

plt.figure(figsize=(12, 8))
plt.barh(product_order_summary['Product Name'], product_order_summary['Order Quantity'], color='blue')
plt.xlabel('Order Quantity')
plt.ylabel('Product Name')
plt.title('Order Quantity by Product Name')
plt.gca().invert_yaxis() 
plt.show()
```

## Sum of Order Quantity by Sales Channel

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Sample Dataset.xlsx'  

df = pd.read_excel(file_path)

summary = df.groupby('Sales Channel')['Order Quantity'].sum()

plt.figure(figsize=(8, 8))
summary.plot.pie(autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral', 'lightgoldenrodyellow'])
plt.ylabel('')
plt.title('Order Quantity by Sales Channel')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


my_data_set = 'Sample Dataset.xlsx'
orders_df = pd.read_excel(my_data_set, sheet_name='Sales Orders Sheet') # reading the data 


orders_df['Final Amount'] = orders_df['Order Quantity'] * orders_df['Unit Cost'] # Finding [final_amount = order_quantiy * unit_cost] 


team_sales = orders_df.groupby('SalesTeam Code')['Final Amount'].sum() # groupBY Sales team code and final amount 


# Here , i'm plot the graph 
team_sales.plot(kind='bar', figsize=(10, 6), color='green')
plt.title('Total Sales Amount by SalesTeam Code')
plt.xlabel('SalesTeam Code')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('team_sales_bar_graph.png')
plt.show()
```
